```python
import numpy as np
import pandas as pd
```

```python
genres = pd.read_csv('genres_v2.csv')
playlists = pd.read_csv('playlists.csv.xls')
```

```python
genres
```

```python
genres_song = genres[~genres['song_name'].isna()]
genres_song
```

```python
genres_song.isnull().sum()
```

```python
genres_song.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
song_features = genres_song.groupby('song_name').mean()
song_features
```

```python
song_features.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
norm_song_features = sc.fit_transform(song_features)
norm_song_features
```

```python
row = song_features.index
col = song_features.columns
df = pd.DataFrame(index = row, columns = col, data = norm_song_features)
df
```

```python
#correlation
```

```python
df_corr = df.corr()
sim = df @ df_corr
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import cosine

from scipy.sparse import csr_matrix

matrix = csr_matrix(sim.values)
matrix
```

```python
sim = cosine_similarity(df)
sim
```

```python
sim.shape
```

```python
sim_df = pd.DataFrame(data = sim, index = row, columns = row)
sim_df
```

```python
np.fill_diagonal(sim_df.values, 0)
```

```python
sim_df
```

```python
def find_correlation(df, song_name): 
    similar_to_song = df.corrwith(df[song_name])
    similar_to_song = pd.DataFrame(similar_to_song, columns = ['Correlation'])
    similar_to_song = similar_to_song.sort_values(by = 'Correlation', ascending = False)
    return similar_to_song
find_correlation(sim_df, '"Enough."')
```

```python
def find_similarity(df, song_name): 
    df2 = df[[song_name]].sort_values(by = song_name, ascending = False)
    df2.rename(columns = {song_name : 'song_recommendations'}, inplace = True)
    return df2.head(10)
find_similarity(sim_df, '"Enough."')
```
